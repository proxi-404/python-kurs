## Sortierverfahren
Unter einem Sortierverfahren versteht man einen Algorithmus, der dazu dient eine Array oder ähnliches zu sortieren. Dabei muss es sich nicht immer um Zahlen handeln. Zum Beispiel kann auch eine Liste Alphabetisch sortiert werden. 
### Bubble Sort 
#### Funktionsweise
Die Funktionsweise des Bubble-Sort Algorithmus ist im Prinzip recht einfach. Die gegebene Liste wird von rechts nach links durchlaufen. Dabei wird das aktuelle Element immer mit dem rechten Nachbarn verglichen. Wenn der nachbar kleiner als das aktuelle Element ist, werden die Elemente vertauscht, falls nicht, wird das nächste Element betrachtet. Wenn die Liste einmal durlaufen ist, wird die Liste erneut durchlaufen, bis alle Elemente sortiert sind.





### Merge Sort
Mergesort ist ein schwierigerer Sortieralgorithmus, der nach dem Divide and Conquer Prinzip arbeitet. Das Prinzip bedeutet, dass ein Problem immer wieder rekursiv in kleinere Teilprobleme zelegt werden, bis diese besser gelöst werden können. So arbeitet auch Mergesort. Die unsortierte Liste wird in kleinere Teillisten zerlegt. Diese kleinen Teillisten werden dann sortiert und dann wieder in richtiger Reihenfolge zusammengefügt (merge).

## Graphalgorithmen

### Einführung in Graphen
Ein Graph in der Graphentheorie ist eine Struktur, ist eine Menge von Objekten und deren Verbindungen zwischen diesen repräsntiert. Ein Graph besteht dabei dann aus Knoten. Die Verbindungen zwischen den Knoten werden Kanten genannt. 
Um diese Beschreibung deutlicher zu machen ein paar Beispiele wie man Graphen darstellen kann: 
- Das Schienennetz kann man als Graphen darstellen, wobei die Schienenverbindungen zwischen den Bahnhöfen die Kanten darstellen und die Bahnhöfe die Knoten. 
- Das Straßennetz: Die Straßen sind die Kanten und die Abfahrten, Kreuzungen und Abbiegungen sind die Knoten. 
- etc

### BFS (Breadth First Search)

In manchen Fällen ist es nötig einen Graphen zu durchlaufen, das heißt alle Knoten eines Graphen zu besuchen. Das kann zum Beispiel hilfreich sein, um einen bestimmten Knoten mit bestimmmten Eigenschaften zu suchen. Dazu gibt es zwei bekannte Algorithmen. BFS und DFS. Hier in diesem Abschnitt wird zuerst Breadth First Search vorgestellt. 

#### Funktionsweise:
Grundlegend wird der Graph erst in besuchte Knoten und nicht besuchte Knoten aufgeteilt. 

BFS untersucht dabei erst alle Knoten, die eine Distanz von eins vom Startknoten aus haben. Dann alle, die die Distanz zwei haben. Das wird solange wiederholt, bis man alle Knoten einmal besucht hat. Dabei werden immer nur Knoten betrachtet, die noch nicht besucht wurden

Hier der Algorithmus einmal als Pseudocode:
```
create a queue Q 
mark v as visited and put v into Q
while Q is non-empty
    remove the head u of Q
    mark and enqueue all neighbors of u
```

### DFS (Depth First Search)

Depth First Search ist ebenfalls ein bekannter Algorithmus zum durchlaufen eines Graphen  

#### Funktionsweise

Wie auch bei BFS wird der Graph zuerst in besuchte und nicht besuchte Knoten unterteilt. 

In dem Algorithmus werden folgende Schritte durchlaufen: 
1. Lege den Startknoten auf einen Stack ab.
2. Nehme den das erste Item von dem Stack und füge es zu der Liste der besuchten Knoten hinzu.
3. Füge alle zu dem gerade besuchten Knoten adjazenten Knoten von oben auf den Stack hinzu, sofern diese noch nicht besucht worden sind.
4. Wiederhole 2. und 3. solange bis der Stack leer ist.

Adjazent bedeutet einfach nur, dass die beiden Knoten direkt über eine Kante verbunden sind.

Es wird hier zwar von einem Stack gesprochen, aber der DFS Algorithmus lässt sich auch mit Hilfe von Rekursion lösen.
