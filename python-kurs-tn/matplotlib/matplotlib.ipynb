## Figure und Axes Objekte
Bisher haben wir einfach `plt.some_command` aufgerufen. `plt` ist dabei nur die Angabe, wo der Befehl hinterlegt ist. Deutlich mehr Optionen erhält man, wenn man mit mit `figure` und `axes` Objekten arbeitet. \
Die `figure` legt allgemeine Parameter wie z.B. die Größe fest. Das allermeiste stellt man für die Axen ein. Ein Grundgedanke dabei ist, dass eine figure viele subplots enthalten kann, die alle eine eigene Achse darstellen.

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.suptitle("suptitle")
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot([i for i in range(-10, 10)], [i**2 for i in range(-10, 10)])
ax1.set_title("ax1 title") # same commands as before but prefix set_
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot([i for i in range(-10, 10)], [i**3 for i in range(-10, 10)])
ax2.set_title("ax2 title") # same commands as before but prefix set_

# nice layout
fig.set_size_inches(5, 10, forward=True)
fig.tight_layout(h_pad = 3)
fig.subplots_adjust(top=0.9)
```

## Beschriftungen
Beschriftungen sind nicht allzu kompliziert, weisen aber eine kleine Tücke bezüglich der Platzierung auf. Man kann sowohl Koordinaten bezüglich der Achsen als auch bezüglich des gesamten Plots angeben (vgl. `#annotation` und `#text under plot`)

```python
import matplotlib.pyplot as plt
import numpy as np

a = [i for i in range(10)]
b = [i**2 - 5 for i in a]
 
fig, ax = plt.subplots() # second way of creating subplots

ax.plot(a, b, "b-")
ax.plot(a, b, "ro", label="Induzierter Strom") #rote Punkte
ax.grid(True)
ax.set_xlabel("t [s]")
ax.set_ylabel("I [mA]")
ax.legend(loc="upper left")
ax.text(0, 0.75, "some text")

# --- text and annotations ---
bb_props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

# arrow with text
ax.text(2.5, 2, "Direction",ha="center", va="center", rotation=45, size=15,
            bbox=dict(boxstyle="rarrow,pad=0.3", fc="lightblue", ec="steelblue", lw=2))

# grid with coordinates
for x in np.linspace(0, max(a), 3):
    for y in np.linspace(0, max(b), 3):
        plt.text(x,y, f"({int(x)},{int(y)})") # text is placed at coordinates (x,y)

#annotation 
ax.annotate("comment", xy=(1,0), xytext=(1.5,9), fontsize=20, \
    arrowprops=dict(facecolor="black", shrink=0.005), bbox=bb_props)

# text under plot
fig.text(0.5, -0.25, r"Some more information, eg. $\chi^2$"+"\n\n Have fun adding more features.",\
         transform=fig.transFigure, bbox=bb_props, ha="center") # transform=fig.transFigure -> lower left corner (0,0) upper right corner (1,1)
```
